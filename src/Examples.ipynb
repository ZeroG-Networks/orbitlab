```python
from org.hipparchus.geometry.euclidean.threed import Vector3D

# Example problems from:
# "Introduction to Space Dynamics" by William Tyrrell Thomson
# ISBN: 978-0-486-65113-2

r = Vector3D(3.0, 2.0, 2.0)

import orbitlab
orbitlab.unit_vector(r)
orbitlab.axes_angles(r)

r1 = Vector3D(4.0, -3.0, 1.0)
r2 = Vector3D(-2.0, 2.0, 3.0)
r1.angle(r1, r2)  # Angle between r1 and r2
r1.crossProduct(r2)
```
